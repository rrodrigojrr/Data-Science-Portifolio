###Weibull distribution fitting

The scale and shape parameters of weibull distribution was estimate minimize the Likelihood function. 


```python

def fitweibull(df):
   from scipy.stats import exponweib
   from scipy.optimize import fmin
   def optfun(theta):
      return -np.sum(np.log(exponweib.pdf(x, 1, theta[0], scale = theta[1], loc = 0)))
   df = df.replace(0, np.nan)
   df = df.dropna(how='all', axis=0)
   x = df.sort_values()
   logx = np.log(x)
   shape = 1.2 / np.std(logx)
   scale = np.exp(np.mean(logx) + (0.572 / shape))
   return fmin(optfun, [shape, scale], xtol = 0.01, ftol = 0.01, disp = 0)

def weib_density(x,shape,scale):
    return (shape / scale) * (x / scale)**(shape - 1) * np.exp(-(x / scale)**shape)
```

###Plot

```python
params_10 = fitweibull(wind['Wind Speed (km/h)'])
params_30 = fitweibull(wind['Wind Speed in 30m (km/h)'])
params_50 = fitweibull(wind['Wind Speed in 50m (km/h)'])
x = np.arange(0,50,0.01)

fig = plt.figure(figsize=(8,5))
ax = fig.subplots()
ax.plot(x,weib_density(x,params_10[0],params_10[1]), label = '10 meters')
ax.plot(x,weib_density(x,params_30[0],params_30[1]), label = '30 meters')
ax.plot(x,weib_density(x,params_50[0],params_50[1]), label = '50 meters')
plt.title(f'Weibull distributions density of wind in Szeged - Hungria')
plt.legend(title='Wind height:')
plt.xlabel('Wind velocity')
plt.ylabel('Density')




```
